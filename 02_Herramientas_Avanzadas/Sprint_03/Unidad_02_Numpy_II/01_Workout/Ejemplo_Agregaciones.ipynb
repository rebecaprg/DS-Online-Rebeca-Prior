##  Ejemplo Funciones Agregacion: Presidentes USA


### Cul es la estatura media de los presidentes de EEUU?
  


Las funciones de agregacin de Numpy pueden resultar de gran utilidad a la hora de resumir sets de datos. En este ejemplo vamos a considerar las diferentes estaturas de los presidentes de Estados Unidos. Tienes estos datos disponibles en el SV *president_heights.csv*.

Utilizaremos el paquete Pandas para leer los datos.

```python
import numpy as np
import pandas as pd

data = pd.read_csv('./data/president_heights.csv')
heights = np.array(data['height(cm)'])
print(heights)
```

Ahora que tenemos los datos en un array podemos calcular una serie de estadsticos

```python
print("Mean height:       ", heights.mean())
print("Standard deviation:", heights.std())
print("Minimum height:    ", heights.min())
print("Maximum height:    ", heights.max())
```

Fjate que en cada operacin de agregacin, estamos reduciendo el array entero a un nico valor, el cual nos da informacin sobre cmo se distribuyen los valores.

Vemos que la altura mediana de los presidentes roza el metro ochenta.

Podemos ver tambin los cuantiles:

```python
print("25th percentile:   ", np.percentile(heights, 25))
print("Median:            ", np.median(heights))
print("75th percentile:   ", np.percentile(heights, 75))
```
